Converting data in 'release_date' to datetime object:

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df.info()
```

```python
df.describe()
```

Create new columns year, month from tn_movie_budgets release_date

```python
import datetime
```

```python
df['year'], df['month'] = df['release_date'].apply(lambda x: x.year), df['release_date'].apply(lambda x: x.month)
df
```

```python
df['production_budget_2'] = df['production_budget'].apply(lambda x: x[1:])
df['domestic_gross_2'] = df['domestic_gross'].apply(lambda x: x[1:])
df['worldwide_gross_2'] = df['worldwide_gross'].apply(lambda x: x[1:])
```

```python
df['genres2'] = df['genres'].str.split(',')
```

```python
len(df['genres2'][1])
```

```python
df['production_budget_2'] = df['production_budget_2'].str.replace(',', '')
df['domestic_gross_2'] = df['domestic_gross_2'].str.replace(',', '')
df['worldwide_gross_2'] = df['worldwide_gross_2'].str.replace(',', '')
```

```python
df['production_budget_2'] = df['production_budget_2'].astype(int)
df['domestic_gross_2'] = df['domestic_gross_2'].astype(int)
df['worldwide_gross_2'] = df['worldwide_gross_2'].astype(int)
```

```python
df.head()
```

```python
df.loc[df['start_year'] != df['year']].sort_values(by = ['worldwide_gross_2'], ascending = False)
```

```python
df = df.drop(columns = ['production_budget', 'domestic_gross', 'worldwide_gross', 'tconst.1'])
```

```python
df['profit'] = (df['worldwide_gross_2'] - df['production_budget_2'])/df['production_budget_2']
```
